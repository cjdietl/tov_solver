```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols = ['Radius','Mass']
mr1 = pd.read_csv('massradius1.00E+12.dat',sep='\t',header=2,names=cols).reset_index()
mr2 = pd.read_csv('massradius2.00E+12.dat',sep='\t',header=2,names=cols).reset_index()
mr3 = pd.read_csv('massradius3.00E+12.dat',sep='\t',header=2,names=cols).reset_index()
sol_to_earth = 3.00273E-6
mr1['Mass'] /=  sol_to_earth
mr2['Mass'] /=  sol_to_earth
mr3['Mass'] /=  sol_to_earth
```

```python
fig, ax = plt.subplots(figsize=(6,8))
#plt.rcParams.update({'font.size': 22})
mr1.plot(ax=ax,x='Radius',y='Mass',label='3 TeV')
mr2.plot(ax=ax,x='Radius',y='Mass',label='1 TeV')
mr3.plot(ax=ax,x='Radius',y='Mass',label='2 GeV')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylim([1E-8,1E0])
ax.set_xlim([0.01,200])
ax.set_ylabel('$M/M_{\mathrm{Earth}}$',fontsize=20)
ax.set_xlabel('Radius (cm)',fontsize=20)
fig.tight_layout()
fig.savefig('MR.png')
```
